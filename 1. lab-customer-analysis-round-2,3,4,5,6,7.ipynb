```python
#----------------------------------------------------------- 1. X-y SPLIT ---------------------------------------------------#

y = data_2['total_claim_amount']
X = data_2.drop('total_claim_amount', axis=1)

#---------------------------------------------------------- 2. TRAIN-TEST-SPLIT ---------------------------------------------#

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)

#---------------------------------------------------- 3. CATEGORICAL AND NUMERICAL SPLIT ------------------------------------#

X_train_num = X_train.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)
X_test_num = X_test.select_dtypes(np.number)
X_test_cat = X_test.select_dtypes(object)

#-------------------------------------------------- 4. PRE-PROCESSING CATEGORICAL VARIABLES -----------------------------------#

#------------------------------ c) One hot encoding -------------------------------------------------------------------------#

### Train encoder on train set
# import
from sklearn.preprocessing import OneHotEncoder
# define and train encoder
encoder = OneHotEncoder(drop='first').fit(X_train_cat) 
# define columns
columnsOHEtrain = encoder.get_feature_names(input_features = X_train_cat.columns)
columnsOHEtest = encoder.get_feature_names(input_features = X_test_cat.columns)

### Onehotencode train set
X_train_categorical_OHE = pd.DataFrame(encoder.transform(X_train_cat).toarray(),columns=columnsOHEtrain)

### Onehotencode test set
X_test_categorical_OHE = pd.DataFrame(encoder.transform(X_test_cat).toarray(),columns=columnsOHEtest)

#----------------------------------------------------- 5. PRE-PROCESSING NUMERICAL VARIABLES ----------------------------------#

#-------------------------------- c) Scaling continuous variables: -----------------------------------------------------------#

### import necessary libraries
import numpy as np
from sklearn.preprocessing import MinMaxScaler

### define scaler
scaler = MinMaxScaler()

### train transformer
scaler.fit(X_train_num)

### scale train set
X_train_num_columns = X_train_num.columns
X_train_num_scaled = pd.DataFrame(scaler.transform(X_train_num), columns= X_train_num_columns) #same thing in 1 line: X_train_scaled = scaler.fit_transform(X_train)

#### scale test set
X_test_num_columns = X_test_num.columns
X_test_num_scaled = pd.DataFrame(scaler.transform(X_test_num), columns = X_test_num_columns)

#------------------------------- 6. CONCATENATE PROCESSED CATEGORICAL AND NUMERICAL DATA --------------------------------------#

### reset index and drop old index
X_train_categorical_OHE = X_train_categorical_OHE.reset_index(drop=True)
X_train_num_scaled = X_train_num_scaled.reset_index(drop=True)

### concatenate
X_train_processed = pd.concat([X_train_num_scaled, X_train_categorical_OHE], axis=1)

### reset index and drop old index
X_test_categorical_OHE = X_test_categorical_OHE.reset_index(drop=True)
X_test_num_scaled = X_test_num_scaled.reset_index(drop=True)

### concatenate
X_test_processed = pd.concat([X_test_num_scaled, X_test_categorical_OHE], axis=1)
```
